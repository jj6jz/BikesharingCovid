```python
# trying to calculate distance
# https://en.wikipedia.org/wiki/Haversine_formula
# https://towardsdatascience.com/exploring-bike-share-data-3e3b2f28760c
# turns out our start and end lat/long the same in VA_clean2 file. 
```

```python
import numpy as np
def haversine(lat1, lon1, lat2, lon2):
    lon1, lat1, lon2, lat2 = map(np.radians ,[lon1, lat1, lon2, lat2])
    h = np.sin((lat2-lat1)/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin((lon2-lon1)/2.0)**2
    miles = 3959 * (2 * np.arcsin(np.sqrt(h))) 
    return miles
```
